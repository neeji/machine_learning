```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("./linear_regression/cities.csv")
df
```

```python
%matplotlib inline
plt.xlabel("1980",fontsize=20)
plt.ylabel("1990",fontsize=20)
plt.scatter(df[['Pop 1980']],df[['Pop 1990']],color='red',marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Pop 1980']],df[['Pop 1990']])
```

```python
reg.intercept_
reg.coef_
```

```python
reg.predict(np.array([[17500]]))
```

```python
# %matplotlib inline
plt.xlabel("1980",fontsize=20)
plt.ylabel("1990",fontsize=20)
plt.scatter(df[['Pop 1980']],df[['Pop 1990']],color='red',marker = '+')
plt.plot(df[['Pop 1980']],reg.predict(df[['Pop 1980']]),color='blue')
```
